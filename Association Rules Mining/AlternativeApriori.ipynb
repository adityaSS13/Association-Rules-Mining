## Apriori algorithm

### 1.apriori using built-in function

it took 1 minut and 12 seonds to run this verison of arioti algorithm

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
transactions = data.values.tolist()
transactions = [[str(item) for item in transaction] for transaction in transactions]

# Use TransactionEncoder to transform the transaction data into a one-hot encoded format
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)

# Convert the one-hot encoded data back to a DataFrame
onehot_df = pd.DataFrame(te_ary, columns=te.columns_)

# Apply Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(onehot_df, min_support=0.01, use_colnames=True)

# Apply association rule mining
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8)

# Display the frequent itemsets and association rules
print("Frequent Itemsets:")
print(frequent_itemsets)

# print("\nAssociation Rules:")
# print(rules)

```

### 2 My implementation for apriori algorithm (Fk-1 * Fk-1 method for candidate generation)

it took 45 minutes to run this version of apriori agorithm



```python
def create_set(data):
    result = []

    # Assuming all values are either strings or floats
    values = data.values

    for row in values:
        row_set = set(value for value in row if pd.notna(value) and not str(value).replace('.', '', 1).isdigit())
        result.append(row_set)

    return result
new_data = create_set(data)
```

```python
from itertools import combinations


def calculate_support(data, itemset):
    count = 0
    for transaction in data:
        if set(itemset).issubset(transaction):
            count += 1
    return count

def generate_candidate1(L,k): # Fk1  Fk1 method
    Ck = []
    for itemset1 in L[k - 2]:
        for itemset2 in L[k - 2]:
            union_set = itemset1.union(itemset2)
            if len(union_set) == k and union_set not in Ck:
                Ck.append(union_set)
    return Ck

def generate_candidate2(L,k): # Fk1  F1 method
    Ck = []
    for itemset1 in L[k - 2]:
        for itemset2 in L[0]:
            union_set = itemset1.union(itemset2)
            if len(union_set) == k and union_set not in Ck:
                Ck.append(union_set)
    return Ck



def apriori(data, min_support, min_confidence):
    items = list(set(item for transaction in data for item in transaction))
    C1 = [{item} for item in items]
    L1 = [item for item in C1 if calculate_support(data, item) >= min_support]

    L = [L1]
    k = 2
    while True:
        Ck = generate_candidate1(L,k)
        print("Numebr of candidates in C"+str(k)+":"+str(len(Ck)))
        # for i in Ck:
        #     print(str(list(i)))

        Lk = []
        for candidate in Ck:
            if calculate_support(data, candidate) >= min_support:
                Lk.append(candidate)
        print(Lk)

        if not Lk:
            break

        L.append(Lk)
        k += 1

    frequent_itemsets = [item for sublist in L for item in sublist]

    for itemset in frequent_itemsets:
        if len(itemset) > 1:
            for Lk in range(1, len(itemset)):
                for subset in combinations(itemset, Lk):
                    remain = itemset.difference(subset)
                    if calculate_support(data, itemset) >= min_support and \
                            calculate_support(data, itemset) / calculate_support(data, subset) >= min_confidence:
                        print(f"Rule: {subset} => {remain} (Confidence: {calculate_support(data, itemset) / calculate_support(data, subset)})")



min_support = 6000
min_confidence = 0.8
apriori(new_data, min_support, min_confidence)
```
